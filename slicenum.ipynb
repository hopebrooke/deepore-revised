```python
def visualize(n):       # n - num of slices in each dir
  # read in first entry
  entry = dpr.readh5slice(dataset, 'X', [0])
  # display midslices
  dpr.showentry(entry, n)
  # get properties
  properties = dpr.readh5slice(dataset, 'Y', [0])
  # show and save properties
  dpr.prettyresult(properties, 'sample_gt.txt', units='px')

visualize(1)    # 3 slices - 1 in each dir
visualize(2)    # 6 slices - 2 in each dir
visualize(3)    # 9 slices - 3 in each dir
```

```python
#----- RETRAIN - 3-------
# prepare data
List = dpr.prep(dataset_1, 1)
# split data
TrainList, EvalList, TestList = dpr.splitdata(List)
# train
model=dpr.trainmodel(dataset_1,TrainList,EvalList,retrain=1,epochs=2,batch_size=100,ModelType=3, n=1)
# test
dpr.testmodel(model,dataset_1,TestList, 1)
```
